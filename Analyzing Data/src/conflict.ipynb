```python


# specify the conflict you are interested in
conflict_name = df['conflict']

# filter the dataframe to only include rows where the 'war' column matches the conflict name
conflict_df = df[df['conflict'] == conflict_name]

# extract the unique values in the 'code' column of the filtered dataframe
code_of_war = conflict_df['ccode'].unique()

# print the code(s) of war that belong to the conflict
print(code_of_war)
```

```python

# get the unique conflict names in the 'conflict' column
conflict_names = df['conflict'].unique()

# loop through the conflict names and print the corresponding ccode values
for conflict in conflict_names:
    conflict_codes = df.loc[df['conflict'] == conflict, 'ccode'].unique()
    print(f'Conflict: {conflict}')
    print(f'CCodes: {conflict_codes}')
```

```python


# get the unique conflict names in the 'conflict' column
conflict_names = df['conflict'].unique()

# create an empty list to store the results
results = []

# loop through the conflict names and append the corresponding ccode values to the results list
for conflict in conflict_names:
    conflict_codes = df.loc[df['conflict'] == conflict, 'ccode'].unique()
    ccode_str = ', '.join(str(c) for c in conflict_codes)
    results.append({'Conflict': conflict, 'CCodes': ccode_str})

# create a new dataframe from the results list and display it
results_df = pd.DataFrame(results)
results_df.head()
results_df.to_excel('../data_processed/conflict_code.xlsx')
```

```python


# get the unique conflict names in the 'conflict' column
conflict_names = df['conflict'].unique()

# create an empty list to store the results
results = []

# loop through the conflict names and append the corresponding ccode and country values to the results list
for conflict in conflict_names:
    conflict_df = df.loc[df['conflict'] == conflict]
    conflict_codes = conflict_df['ccode'].unique()
    countries = conflict_df['interven'].unique()
    ccode_str = ', '.join(str(c) for c in conflict_codes)
    country_str = ', '.join(str(c) for c in countries)
    results.append({'Conflict': conflict, 'CCodes': ccode_str, 'Countries': country_str})

# create a new dataframe from the results list and display it
results_df = pd.DataFrame(results)
results_df.head()
results_df.to_excel('../data_processed/conflict_code.xlsx')
```

```python
import matplotlib.pyplot as plt


# group the dataframe by country and count the number of conflicts
country_counts = df.groupby('interven').nunique()['conflict']

# create a bar chart showing the number of conflicts for each country
country_counts.plot(kind='bar', figsize=(10, 5))

# set the chart title and axis labels
plt.title('Number of Conflicts by Country')
plt.xlabel('Country')
plt.ylabel('Number of Conflicts')

# show the chart
plt.show()
```
